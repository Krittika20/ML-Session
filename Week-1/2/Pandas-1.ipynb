```python
import pandas as pd 
import numpy as np
```

### Pandas
<p>pandas. pandas is a fast, powerful, flexible and easy to use open source data analysis and manipulation tool, built on top of the Python programming language.</p>

#### Series
Pandas Series is a one-dimensional labeled array capable of holding data of any type (integer, string, float, python objects, etc.). The axis labels are collectively called index. Pandas Series is nothing but a column in an excel sheet. Labels need not be unique but must be a hashable type. The object supports both integer and label-based indexing and provides a host of methods for performing operations involving the index

```python
# Creating empty series 
ser = pd.Series() 

# simple array 
data = np.array(['pandas','numpy','matplotlib','opencv']) 
ser = pd.Series(data) 
print(ser) 
```

#### DataFrame
Pandas DataFrame is two-dimensional size-mutable, potentially heterogeneous tabular data structure with labeled axes (rows and columns). A Data frame is a two-dimensional data structure, i.e., data is aligned in a tabular fashion in rows and columns. Pandas DataFrame consists of three principal components, the data, rows, and columns<br>
[DataFrame]('https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.html')

```python
#creating DataFrame

# intialise data of lists.
data = {'Name':['Tom', 'nick', 'krish', 'jack'],
        'Age':[20, 21, 19, 18]}
 
# Create DataFrame
df = pd.DataFrame(data)
 
# Print the output.
df
```

### Selection and Deletion
[drop]('https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.drop.html')
#### Column

```python
## adding new column 
df['WorkDone'] = ['Yes','No','No','Yes']
df
```

```python
## selecting Multiple Column
df[['Name','WorkDone']]
```

```python
df.drop(['WorkDone'], axis=1, inplace=True)
df
```

#### Row

```python
## Addition of new Row
newRow = {'Name':"ravi", "Age":21}
newDf = pd.DataFrame(newRow, index=[0])
df = pd.concat([newDf, df]).reset_index(drop=True)
df
```

```python
df.drop(0, inplace=True)
df
```

```python
print(df)
def change(row):
    if row =='Yes':
        return 1
    else:
        return 0


## Apply 
df['WorkDone'] = df['WorkDone'].apply(change)
df
```

```python
#No changes have been made because drop function does not delete the rows from the original dataframe. 
#It creates a copy of the dataframe and returns the updated copy

#To make changes in the original dataframe we need to pass a inplace=True argument. This ensures that the changes are made 
#in the original dataset

data.drop(0, inplace=True)
data.head()
```
